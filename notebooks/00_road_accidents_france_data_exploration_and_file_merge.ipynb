Doing this will enable us to work with a cleaner and more consistent dataset. First because the reporting formats have been more stable in recent years, but also focusing on 2019–2023 makes it easier to detect and handle any data inconsistencies on a year-by-year basis. At the same time, we still benefit from a dataset that is large enough (for example, more than 200,000 rows for df_places alone) to identify meaningful trends and produce insightful visualizations.

## 2. Concatenating the users files 2019 - 2023

```python
# File paths
files = {
    2019: "users_2019.csv",
    2020: "users_2020.csv",
    2021: "users_2021.csv",
    2022: "users_2022.csv",
    2023: "users_2023.csv",
}
# Read, add year column, and store in a list
dfs = []
for year, path in files.items():
    df = pd.read_csv(path, sep = ";", low_memory = False)

    # Ensure id_usager exists and make it all a string
    if "id_usager" not in df.columns:
        df["id_usager"] = pd.NA
    else:
        df["id_usager"] = df["id_usager"].astype(str)

    # Normalize an_nais
    if "an_nais" in df.columns:
        df["an_nais"] = (
            pd.to_numeric(df["an_nais"], errors = "coerce")  # handles float/int mix
            .round()                                       # turn 1985.0 → 1985
            .astype("Int64")                               # nullable integer dtype
        )

    # Add year column
    df["year"] = year
    dfs.append(df)

# Concatenate all years for users 2019 - 2023
combined_df = pd.concat(dfs, ignore_index = True)

# Check for duplicates
duplicates = combined_df.duplicated()

# Count the number of duplicates
num_duplicates = duplicates.sum()

print("Number of duplicate rows:", num_duplicates)

# Save to a single CSV
combined_df.to_csv("usagers-2019-2023.csv", sep = ";", index = False)
print("Combined file saved as usagers-2019-2023.csv")
```

#### Making one row per accident in the places file

```python
#  Aggregate places per accident into pipe-separated lists (no new columns)

def build_places_agg(df_places):
    key = "Num_Acc"
    non_key_cols = [c for c in df_places.columns if c != key]

    out = pd.DataFrame(index=df_places[key].drop_duplicates().sort_values())

    for col in non_key_cols:
        # keep only (Num_Acc, col), drop NaNs, deduplicate pairs early
        s = df_places[[key, col]].dropna(subset=[col])
        # using strings is safer for join; pyarrow string if available is faster/memory-friendly
        try:
            s[col] = s[col].astype("string[pyarrow]")
        except Exception:
            s[col] = s[col].astype("string")

        s = s.drop_duplicates()  # (Num_Acc, value) pairs

        # group and join
        agg_col = s.groupby(key, sort=False)[col].agg("|".join)

        # align on index
        out[col] = agg_col

    out = out.reset_index()  # bring Num_Acc back as a column
    return out

# Aggregate places
df_places_agg = build_places_agg(df_places)


# Count the number of accidents recorded
unique_count = df_places_agg["Num_Acc"].nunique()
print("The number of accidents recorded in the aggregated places file is:", unique_count)

# Identify columns that contain "|" (meaning multiple values joined)
cols_with_pipe = [
    col for col in df_places_agg.columns
    if df_places_agg[col].astype(str).str.contains(r"\|", na=False).any()
]

print("\nColumns that contain '|' (aggregated from multiple place rows):")
print(cols_with_pipe)

# Show how many rows per column have multiple values
pipe_counts = (
    df_places_agg
      .astype(str)
      .apply(lambda col: col.str.contains(r"\|", na=False).sum())
)
pipe_counts = pipe_counts[pipe_counts > 0]  # only those with at least one "|"

print("\nNumber of rows with '|' per column:")
print(pipe_counts)

df_places_agg.info()
```

```python

import numpy as np

# Work only on rows that contain a "|"
mask = df_places_agg["vma"].astype(str).str.contains(r"\|", na=False)

def fix_vma(val):
    if pd.isna(val):
        return pd.NA

    parts = [p.strip() for p in str(val).split("|") if p.strip() != ""]

    # Try to parse numbers; if non-numeric appears, return NA
    try:
        nums = [float(p) for p in parts]
    except ValueError:
        return pd.NA

    # Remove duplicates, preserve order
    nums = list(dict.fromkeys(nums))

    # Case 1: both values are -1 → return NA
    if len(nums) == 2 and nums[0] == -1 and nums[1] == -1:
        return pd.NA

    # Case 2: two numbers where one is -1 → keep the other
    if len(nums) == 2 and (-1 in nums) and any(n != -1 for n in nums):
        return int([n for n in nums if n != -1][0])

    # Case 3: two valid numbers → mean, rounded to whole number
    if len(nums) == 2 and (-1 not in nums):
        return int(round((nums[0] + nums[1]) / 2))

    # Case 4: general fallback — drop -1, average remaining, rounded
    remaining = [n for n in nums if n != -1]
    if not remaining:
        return pd.NA
    return int(round(sum(remaining) / len(remaining)))

# --- Allow integer assignment into the column
df_places_agg["vma"] = df_places_agg["vma"].astype("object")

# Apply only to rows that contain '|'
df_places_agg.loc[mask, "vma"] = df_places_agg.loc[mask, "vma"].map(fix_vma)

# Convert column to numeric (nullable Int64)
df_places_agg["vma"] = pd.to_numeric(df_places_agg["vma"], errors="coerce").astype("Int64")

# Optional: verify
print("Rows with '|' in vma after fix:",
      df_places_agg["vma"].astype(str).str.contains(r"\|", na=False).sum())
print("Rows where vma is NA:", df_places_agg["vma"].isna().sum())

# show counts of the unique values in vma
display(df_places_agg['vma'].value_counts())
```

## 9. Merging the characteristics, places, vehicules and users files

```python

# Keep 'year' only in characteristics
for df in (df_places_agg, df_vehicules, df_users):
    df.drop(columns="year", inplace=True, errors="ignore")


# Merge for user-level grain (one row per person)
full_data = (
    df_users
      .merge(df_vehicules,
             on=["Num_Acc", "id_vehicule", "num_veh"],
             how="left")
      .merge(df_characteristics,
             on="Num_Acc",
             how="left")
      .merge(df_places_agg,
             on="Num_Acc",
             how="left")
)

# Save
full_data.to_csv("road_accidents_france_2019_2023.csv", sep=";", index=False)
print("Combined file saved as road_accidents_france_2019_2023.csv")
```
